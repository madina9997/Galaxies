```python
import numpy as np 
import pandas as pd

# curve-fit() function imported from scipy 
from scipy.optimize import curve_fit 

from matplotlib import pyplot as plt 
```

```python
def test(x, a, b): 
    #my hypothesis its Log
	return a * np.log(b * x) 
```

```python
def Curve(x,y):
    zipped=zip(x,y)
    l=list(sorted(zipped, key=lambda x: x[0]))
    lst=list(zip(*l))
    x=np.array(list(lst[0]))
    y=np.array(list(lst[1]))

    # curve_fit() function takes the test-function 
    # x-data and y-data as argument and returns 
    # the coefficients a and b in param and 
    # the estimated covariance of param in param_cov 
    param, param_cov = curve_fit(test, x, y) 

    print("Funcion coefficients:") 
    print(param) 
    print("Covariance of coefficients:") 
    print(param_cov) 

    # ans stores the new y-data according to 
    # the coefficients given by curve-fit() function 
    ans = (param[0]*(np.log(param[1]*x))) 

    plt.plot(x, y, 'o', color ='red', label ="data") 
    plt.plot(x, ans, '--', color ='blue', label ="optimized data") 
    plt.legend() 
    plt.title('Dependency of logMtot on Ntot')
    plt.xlabel('Ntot')
    plt.ylabel('logMtot')
    plt.show() 
```

```python
List2 = pd.read_csv("grlist_2.csv")
List2=List2[List2.Ntot >30]
List2.dropna(subset=['logMtot'])
```

```python
ListS = pd.read_csv("grlist_s.csv")
ListS=ListS[ListS.Ntot >30]
ListS.dropna(subset=['logMtot'])
```

Хитмап для иллюстрации кол-ва групп галактик с отображаемыми координатами

```python
def HeatmapForCatalog(List2):
    N_bins = 100

    # Construct 2D histogram from data using the 'plasma' colormap
    # plt.hist2d(List2['Ntot'],[math.exp(i) for i in List2['logMtot']], bins=N_bins, normed=False, cmap='plasma')
    plt.hist2d(List2['Ntot'], List2['logMtot'], bins=N_bins, normed=False, cmap='plasma')

    # Plot a colorbar with label.
    cb = plt.colorbar()
    cb.set_label('Number of groups ')

    # Add title and labels to plot.
    plt.title('Heatmap dependency of logMtot on Ntot')
    plt.xlabel('Ntot')
    plt.ylabel('logMtot')

    # Show the plot.
    plt.show()
```

```python
HeatmapForCatalog(List2)
```

```python
List2_reduced=List2[List2.Ntot<100]
HeatmapForCatalog(List2_reduced)
```

```python
HeatmapForCatalog(ListS)
```

```python
ListS_reduced=ListS[ListS.Ntot<100]
HeatmapForCatalog(ListS_reduced)
```
