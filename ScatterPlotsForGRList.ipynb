```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

grList2 = pd.read_csv("grList2.csv")
grListS = pd.read_csv("grListS.csv")
gaListS = pd.read_csv("gaListS.csv")
gaList2 = pd.read_csv("gaList2.csv")

# print(len(grListS))
# print(len(grList2))
# print(len(gaListS))
# print(len(gaList2))

merge2=pd.merge(grList2, gaList2, on='iGrID')
merge2Ntot30=merge2[merge2['Ntot']>30]
print(merge2Ntot30.head(5))

mergeS=pd.merge(grListS, gaListS, on='iGrID')
mergeSNtot30=mergeS[mergeS['Ntot']>30]
print(mergeSNtot30.head(5))

```

```python
group=merge2Ntot30.iGrID.unique()
for id in group[:3]:
    print(id)
    el=merge2Ntot30[merge2Ntot30["iGrID"]==id]

    x=np.sqrt(np.power(np.array((el['RAJ2000_x']))-np.array((el['RAJ2000_y'])),2)+
                  np.power(np.array((el['DEJ2000_x']))-np.array((el['DEJ2000_y'])),2)
                  )
    y = np.abs(np.array((el['z_x']))-np.array((el['z_y'])))
    colors = np.random.rand(len(el))
    area = (3) ** 2 
    plt.scatter(x, y, s=area, c=colors, alpha=0.5)
    plt.xlabel('iGrID = %d, distance betw gal and gr center' % (id))
    plt.ylabel('iGrID = %d, diff in redshift' % (id))
    plt.show()
```
